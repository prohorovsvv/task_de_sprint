Импорт библиотек

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Построение распределений

```python
import scipy.stats as sps
x = np.linspace(sps.rayleigh.ppf(0.01),
                sps.rayleigh.ppf(0.99), 100)

r = sps.rayleigh.rvs(size=1000)
plt.hist(r, density=True,  alpha=0.2)
plt.plot(x, sps.rayleigh.pdf(x),lw=4, )
plt.show()
```

# Построение графика на по функции

```python
x = np.arange(-2, 2, 0.001)
y1 = 0.5*((x**2)**(1/3)-((x**4)**(1/3)-(4*x**2)+4)**0.5)
y2 = 0.5*((x**2)**(1/3)+((x**4)**(1/3)-(4*x**2)+4)**0.5)
plt.plot(x,y1)
plt.plot(x,y2)
plt.grid()
plt.show()
```

# Анализ зависимостей

Обработка файла

```python
age=[]
sex=[]
bmi=[]
children=[]
smoker=[]
region=[]
charges=[]

with open ("./dataset_home.txt", mode="r") as file:
  for human in file.readlines():
    age.append(int(human.split(',')[0]))
    if human.split(',')[1] == 'male':
      sex.append(1)
    else:
      sex.append(0)
    bmi.append(float(human.split(',')[2]))
    children.append(int(human.split(',')[3]))
    if human.split(',')[4] == 'yes':
      smoker.append(1)
    else:
      smoker.append(0)

    if human.split(',')[5] == 'northwest':
      region.append(0)
    elif human.split(',')[5] == 'northeast':
      region.append(1)
    elif human.split(',')[5] == 'southeast':
      region.append(2)
    elif human.split(',')[5] == 'southwest':
      region.append(3)
    else:
      region.append(human.split(',')[5])

    charges.append(float(human.split(',')[6]))

age=np.array(age)
sex=np.array(sex)
bmi=np.array(bmi)
children=np.array(children)
smoker=np.array(smoker)
region=np.array(region)
charges=np.array(charges)
```

Соотношение возраста и стоимости страховки

```python
plt.scatter(age, charges)
plt.xlabel('Возраст')
plt.ylabel('Стоимость страховки')
plt.grid(True)
plt.show()
```

Зависимость пола и стоимости страховки

```python
plt.bar(sex, charges)
plt.xlabel('Пол')
plt.ylabel('Стоимость страховки')
plt.show()
```

Зависимость курения и стоимсоти страховки

```python
plt.bar(smoker, charges)
plt.xlabel('Курение')
plt.ylabel('Стоимость страховки')
plt.show()
```
